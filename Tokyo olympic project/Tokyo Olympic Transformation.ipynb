```python
## Get all coaches who coached more than one event.

coaches.groupby(by= 'Name')['Event'].\
    agg(Events_coached ='count').reset_index().\
        query('Events_coached > 1')
```

```python
## Join Athletes with Teams using country + discipline to identify which teams athletes were part of. INCORRECT

athletes_teams = pd.merge(
    athletes, 
    teams, 
    left_on=['Country', 'Discipline'], 
    right_on=['Country', 'Discipline'], 
    how='inner'
    )
```

```python
athletes_teams
```

```python
## Aggregate the number of athletes per country and join with Medals to compute medals per athlete ratio.

athlete_count = athletes.groupby(
    by='Country')['PersonName'].\
        agg(total_athletes = 'count')\
            .reset_index()
```

```python
athletes_medals = pd.merge(medals, athlete_count, left_on='TeamCountry', right_on='Country', how='inner')
```

```python
athletes_medals
```

```python
athletes_medals['medals_per_athlete'] = (athletes_medals['Total'] / athletes_medals['total_athletes']).round(2)
```

```python
athlete_medal_ratio = athletes_medals[['TeamCountry', 'Total', 'total_athletes', 'medals_per_athlete']]
```

```python
athlete_medal_ratio.sort_values(by='medals_per_athlete', ascending=False)
```

```python
## Identify the top 5 countries with the highest medal efficiency

top5_countries = athlete_medal_ratio. \
    sort_values(by='medals_per_athlete', ascending=False). \
    head(5)
```

```python
top5_countries['rank'] = top5_countries['medals_per_athlete']. \
    rank(ascending=False). \
        astype(int)
```

```python
top5_countries[['TeamCountry', 'medals_per_athlete', 'rank']]
```
